## <span style="color:#2F749F;"><strong>Ejercicio 1. Limpieza e imputación de un conjunto de datos mixto</strong></span>

1. Crea el DataFrame base **df** ejecutando el siguiente código:
    ```python
    df = pd.DataFrame({
        'Nombre': ['Ana', 'Luis', 'Pedro', None, 'Marta', 'Luis', 'Sofía'],
        'Edad': [25, np.nan, 35, 29, -5, 25, None],
        'Ciudad': ['Bogotá', 'Medellín', None, 'Medellín', 'Cali', 'Bogotá', 'Cali'],
        'Ingreso': [3500, 4800, np.nan, 5200, 5100, np.nan, 4700],
        'FechaIngreso': ['2023-01-01', '2023-01-05', None, '2023-01-10', '2023-01-12', None, '2023-01-15']

2. En un nuevo notebook, aplica las etapas del proceso de limpieza e imputación de datos que consideres necesarias sobre el DataFrame df. Ten encuenta incluir:
    - Código correctamente comentado y ejecutado.
    - Explicaciones breves de cada paso.
    - Resultados visuales o estadísticos que evidencien las transformaciones.

````python
import pandas as pd
import numpy as np

# Creación del DataFrame
df = pd.DataFrame({
    'Nombre': ['Ana', 'Luis', 'Pedro', None, 'Marta', 'Luis', 'Sofía'],
    'Edad': [25, np.nan, 35, 29, -5, 25, None],
    'Ciudad': ['Bogotá', 'Medellín', None, 'Medellín', 'Cali', 'Bogotá', 'Cali'],
    'Ingreso': [3500, 4800, np.nan, 5200, 5100, np.nan, 4700],
    'FechaIngreso': ['2023-01-01', '2023-01-05', None, '2023-01-10', '2023-01-12', None, '2023-01-15']
})
````

````python
df
````

````python
print("Valores nulos por columna:")
print(df.isnull().sum())
````

````python
# Eliminamos filas duplicadas exactas
df = df.drop_duplicates().reset_index(drop=True)
print(f"Registros después de eliminar duplicados: {len(df)}")
````

````python
# Convertimos valores negativos en NaN para luego imputarlos correctamente
df.loc[df['Edad'] <= 0, 'Edad'] = np.nan

# Verificamos el cambio
df['Edad']
````

````python
# Imputación de Edad e Ingreso con la mediana
df['Edad'] = df['Edad'].fillna(df['Edad'].median())
df['Ingreso'] = df['Ingreso'].fillna(df['Ingreso'].median())

# Imputación de Ciudad con la moda (el primer valor de la serie)
df['Ciudad'] = df['Ciudad'].fillna(df['Ciudad'].mode()[0])

# Imputación de Nombre
df['Nombre'] = df['Nombre'].fillna("Desconocido")

print("Estado de nulos después de la imputación:")
df[['Nombre', 'Edad', 'Ciudad', 'Ingreso']].isnull().sum()
````

````python
# Conversión a datetime
df['FechaIngreso'] = pd.to_datetime(df['FechaIngreso'])

# Imputación de fechas
df['FechaIngreso'] = df['FechaIngreso'].fillna(method='ffill')

# Asegurar que Edad sea un entero
df['Edad'] = df['Edad'].astype(int)
````

````python
print("Dataframe limpio")
df
````

````python
import pandas as pd

# Creación del Dataframe
df = pd.DataFrame({
    'ID': [101, 102, 102, 103, 104, 104, 104],
    'Nombre': ['Ana', 'Luis', 'Luis', 'Marta', 'Carlos', 'Carlos', 'Carlos'],
    'Edad': [25, 30, 30, 29, 40, 40, 41],
    'Ciudad': ['Bogotá', 'Cali', 'Cali', 'Medellín', 'Cali', 'Cali', 'Cali'],
    'FechaRegistro': ['2023-01-01', '2023-01-05', '2023-01-05', '2023-01-10',
                    '2023-01-15', '2023-01-15', '2023-01-16']
})
````

#### **¿Cuál es el total de registros originales?**

````python
total_original = len(df)
print(f"Total de registros originales: {total_original}")
````

#### **¿Cuáles y cuántos son los duplicados exactos?**

````python
# Identificar filas que son copias exactas
duplicados_exactos = df[df.duplicated(keep=False)]
cant_exactos = df.duplicated().sum()

print(f"Cantidad de filas que son duplicados exactos: {cant_exactos}")
print("Registros duplicados exactos:")
print(duplicados_exactos)
````

#### **¿Cuáles y cuántos son los duplicados por varias columnas?**

````python
# Buscamos registros donde el ID y Nombre se repitan
duplicados_columnas = df[df.duplicated(subset=['ID', 'Nombre'], keep=False)]
cant_columnas = df.duplicated(subset=['ID', 'Nombre']).sum()

print(f"Cantidad de registros con ID y nombre repetido: {cant_columnas}")
print("Registros con ID/Nombre duplicado:")
print(duplicados_columnas)
````

#### **¿Cuántos registros quedan después de la limpieza?**

````python
df_limpio = df.drop_duplicates(subset=['ID'], keep='last') # Mantenemos el registro más reciente
print(f"Registros finales: {len(df_limpio)}")
print(df_limpio)
````

## <span style="color:#2F749F;"><strong>Ejercicio 3. Corrección de errores tipográficos o de codificación</strong></span>


1. Crea el DataFrame base **df** ejecutando el siguiente código:
    ```python
    df = pd.DataFrame({
        'Ciudad': ['bogota', 'Bogotá', 'BOGOTA', 'bogotá', 'bogata', 'Bógota', 'BogoTa', 'Cali', 'calí', 'medellín', 'medellin']
    })

2.  En el mismo notebook anterior, para el nuevo dataframe df, estandariza la variable ciudad utilizando python.

````python
import pandas as pd

# Creación del DataFrame con inconsistencias
df = pd.DataFrame({
    'Ciudad': ['bogota', 'Bogotá', 'BOGOTA', 'bogotá', 'bogata', 'Bógota', 'BogoTa', 'Cali', 'calí', 'medellín', 'medellin']
})

print(f"Categorías únicas iniciales: {df['Ciudad'].nunique()}")
print(df['Ciudad'].unique())
````

**Estrategia de Estandarización**

Para resolver esto, seguiremos estos pasos:

1.  **Normalización de mayúsculas:** todo a minúsculas o formato título.
2.   **Eliminación de espacios:**   quitar espacios en blanco al inicio o final (strip).
1.   **Eliminación de tildes:** usar una función para quitar acentos.
2.   **Corrección de errores tipográficos:** reemplazar manualmente valores erróneos (como "bogata").












````python
import unicodedata

# Función para eliminar tildes y caracteres especiales
def normalizar_texto(texto):
    if pd.isna(texto):
        return texto
    # Normalizar a forma NFD (separa la letra del acento) y filtrar los caracteres de combinación
    texto = unicodedata.normalize('NFD', texto)
    texto = texto.encode('ascii', 'ignore').decode("utf-8")
    return texto.lower().strip()

# Aplicamos la normalización básica
df['Ciudad_Limpia'] = df['Ciudad'].apply(normalizar_texto)

# Corrección de errores específicos (typos)
# Reemplazamos 'bogata' por la forma correcta
df['Ciudad_Limpia'] = df['Ciudad_Limpia'].replace({'bogata': 'bogota'})

# Finalmente, pasamos a Capitalize para estética (Bogota, Cali, Medellin)
df['Ciudad_Final'] = df['Ciudad_Limpia'].str.capitalize()

print("\nCategorías únicas finales:")
print(df['Ciudad_Final'].unique())
````

## <span style="color:#2F749F;"><strong>Ejercicio 4. Identificación de mecanismos de datos faltantes (MCAR, MAR, MNAR)</strong></span>
1. Se tiene una base de datos de estudiantes universitarios con las siguientes variables: , Edad, Género, Promedio académico, Ingreso familiar y Horas de estudio por semana. Sin embargo, hay valores faltantes en algunas variables.
2. En los siguientes escenarios, clasifica el tipo de dato faltante (MCAR, MAR o MNAR) y justifica tu respuesta en máximo 2 líneas.
   - **Escenario 1.**  Al analizar los datos, se observa que los valores faltantes en Ingreso familiar aparecen de forma completamente aleatoria, sin relación con ninguna otra variable (ni edad, ni género, ni promedio, etc.).
   - **Escenario 2.** Se detecta que los valores faltantes en Horas de estudio ocurren principalmente en estudiantes con promedios bajos.
   - **Escenario 3.** Se detecta que los valores faltantes en Horas de estudio ocurren principalmente en estudiantes con promedios bajos.
   - **Escenario 4.** Se observa que los valores faltantes en Promedio académico aparecen con mayor frecuencia en estudiantes de cierto género.
   - **Escenario 5.** Algunos estudiantes con muy pocas horas de estudio no reportan esta variable porque consideran que el valor es “demasiado bajo” para registrarlo.
   

*   **Escenario 1**

**Clasificación:** MCAR (Missing Completely At Random). La falta de datos no depende de ninguna variable observada ni del valor de la variable misma; es puramente fruto del azar.

*   **Escenario 2 y 3**

**Clasificación:** MAR (Missing At Random). La probabilidad de que falte el dato depende de otra variable observada en el sistema (el promedio académico), no del azar.

*   **Escenario 4**

**Clasificación:** MAR (Missing At Random). El patrón de ausencia está sistemáticamente relacionado con una característica conocida del estudiante (el género).

*  **Escenario 5**


**Clasificación:** MNAR (Missing Not At Random). El dato falta precisamente por el valor que debería haber sido registrado; existe una decisión consciente o sesgo basado en el dato oculto.

2. Al analizar el dataframe `df`:

3. Realiza la imputación de valores faltantes utilizando KNN y regresión

````python
# Análisis de datos faltantes
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.linear_model import LinearRegression

# Recreación del DataFrame
np.random.seed(42)
n = 30
df = pd.DataFrame({
    'Edad': np.random.randint(18, 60, n),
    'Promedio': np.round(np.random.uniform(2.0, 5.0, n), 2),
    'Horas_estudio': np.random.randint(0, 40, n),
    'Ingreso_familiar': np.random.randint(800000, 8000000, n),
    'Semestres_cursados': np.random.randint(1, 10, n)
})

# Introducción de nulos
df.loc[[2, 5, 8, 15], 'Promedio'] = np.nan
df.loc[[1, 7, 10, 18], 'Horas_estudio'] = np.nan
df.loc[[3, 6, 14], 'Ingreso_familiar'] = np.nan

# Análisis de faltantes
print(df.isnull().mean() * 100)
````

**¿Qué variables tienen mayor proporción de datos faltantes?**

Las variables con mayor proporción de nulos son `Promedio` y `Horas_estudio`, ambas con un 13.33% (correspondiente a 4 registros de los 30 totales). Le sigue `Ingreso_familiar` con un 10% (3 registros). Las variables `Edad` y `Semestres_cursados` están completas (0%).

**¿Observas algún patrón estructurado o parecen aleatorios?**

Los datos parecen seguir un patrón aleatorios. No se observa que la falta de un dato dependa del valor de otra variable (por ejemplo, los nulos en "Ingreso_familiar" ocurren tanto en estudiantes jóvenes como mayores). Es una distribución dispersa diseñada para probar la robustez de los algoritmos de imputación.

````python
# Imputación de valores faltantes usando KNN y regresión
# Copia del DF para KNN
df_knn = df.copy()

# Inicializamos el imputador (usaremos 3 vecinos para un dataset pequeño)
imputer_knn = KNNImputer(n_neighbors=3)
df_knn_imputed = pd.DataFrame(imputer_knn.fit_transform(df_knn), columns=df_knn.columns)

print("Muestra de datos imputados con KNN (Fila 2 - Promedio):")
print(df_knn_imputed.loc[2, 'Promedio'])

# -------- Verificación --------
print()
print("Valores faltantes antes de imputar:")
print(df_knn.isnull().sum())

print("\nValores faltantes después de imputar:")
print(df_knn_imputed.isnull().sum())
````

````python
# Imputación por regreison
df_reg = df.copy()

def imputar_regresion(dataset, columna_objetivo):
    # Separamos filas con y sin nulos
    test_data = dataset[dataset[columna_objetivo].isnull()]
    train_data = dataset.dropna()

    # Columnas predictoras (todas menos la que queremos imputar)
    X_train = train_data.drop(columns=[columna_objetivo])
    y_train = train_data[columna_objetivo]
    X_test = test_data.drop(columns=[columna_objetivo])

    X_test = X_test.fillna(X_train.mean())

    modelo = LinearRegression()
    modelo.fit(X_train, y_train)
    predicciones = modelo.predict(X_test)

    dataset.loc[dataset[columna_objetivo].isnull(), columna_objetivo] = predicciones

# Imputamos las 3 columnas con nulos
for col in ['Promedio', 'Horas_estudio', 'Ingreso_familiar']:
    imputar_regresion(df_reg, col)

print("Muestra de datos imputados con Regresión (Fila 2 - Promedio):")
print(df_reg.loc[2, 'Promedio'])

# Verificación
print()
print("Valores faltantes antes de imputar:")
print(df.isnull().sum())

print("\nValores faltantes después de imputar:")
print(df_reg.isnull().sum())
````

Para determinar si se preserva la naturaleza de los datos, debemos observar lo siguiente:

**1. Con KNN:**

Es excelente para mantener la consistencia local. Al usar vecinos, el valor imputado (3.47) está dentro del rango real de los datos y se basa en perfiles similares.


**2. Con regresión:**

Preserva la relación estadística (correlación) entre variables. Si en la vida real el promedio sube cuando suben las horas de estudio, la regresión mantendrá esa lógica de manera estricta.

**¿Qué variables presentan datos faltantes consecutivos?**

* `Clientes`: 2 rachas de datos faltantes

**¿Por qué este tipo de patrón es relevante en series temporales?**

Porque en una serie temporal, los datos tienen un orden, por lo que hay que tener en cuenta no solo cuántos faltan, sino también cuándo y durante cuánto tiempo.

**¿Qué implicaciones tiene el orden de los datos para este tipo de imputación?**

El orden de los datos es fundamental porque en una serie temporal existe una dependencia cronológica; el valor de hoy suele estar relacionado con el de ayer. Si se altera el orden o se ignora, se pierde la estructura de la serie y se pueden generar datos imputados que no tienen sentido lógico con la tendencia o la estacionalidad del fenómeno.
